```python
#- plot all lines in one plot. Add color/labels for each line
fig = plt.figure(figsize = (12, 6))
plt.plot(unrate.MONTH.head(12), unrate.UNRATE.head(12), c='red', label = '1948' )
plt.plot(unrate.MONTH[12:24], unrate.UNRATE[12:24], c='blue', label = '1949' )
plt.plot(unrate.MONTH[24:36], unrate.UNRATE[24:36], c='green', label = '1950' )
plt.plot(unrate.MONTH[36:48], unrate.UNRATE[36:48], c= 'orange', label = '1951' )
plt.plot(unrate.MONTH[48:60], unrate.UNRATE[48:60], c= 'black', label = '1952' )
plt.legend(loc='upper right')
plt.show()
```
